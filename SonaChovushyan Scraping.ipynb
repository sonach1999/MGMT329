```python
from bs4 import BeautifulSoup
import requests

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36"
}

url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
resp = requests.get(url, headers=headers)
print(resp.status_code)
```

```python
## Title = The Shawshank Redemption
soup = BeautifulSoup(resp.content, 'html.parser')
table = soup.find("div", class_="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-a69a4297-2 bqNXEn cli-title with-margin")
title = table.find("h3", class_="ipc-title__text").text.strip()
title

```

```python
titles_table = soup.find_all("div", class_="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-a69a4297-2 bqNXEn cli-title with-margin")
all_titles = []

for i in titles_table:
    titles = i.text
    all_titles.append(titles)
```

```python
all_titles
```

```python
table_year = soup.find("div", class_="sc-5bc66c50-5 hVarDB cli-title-metadata")
year = table_year.find("span", class_="sc-5bc66c50-6 OOdsw cli-title-metadata-item").text.strip()
year
```

```python
table1 = soup.find_all("span", class_="sc-5bc66c50-6 OOdsw cli-title-metadata-item")
years = []

for i in range(0,len(table1),3):
    info = table1[i].text
    years.append(info)

years
```

```python
Duration = []

for i in range(1,len(table1),3):
    info = table1[i].text
    Duration.append(info)

Duration
```

```python
Age_rating = []

for i in range(1,len(table1),3):
    info = table1[i].text
    Age_rating.append(info)

Age_rating
```

```python
starrate = rating.find("span", class_="ipc-rating-star--rating").text.strip()
starrate
```

```python
import pandas as pd
```

```python
Movies = pd.DataFrame(data={
    "Rank & Title": all_titles ,
    "Year": years,
    "Duration": Duration,
    "Age rating": Age_rating,
    "Star rating": ratings,
    "Number of voters": voters})
Movies.to_csv("./Movies.csv", sep=',',index=False)
Movies
```
